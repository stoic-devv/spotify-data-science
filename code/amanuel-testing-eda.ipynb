```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns 

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
df = pd.read_csv('../data/final-data.csv', sep='~')
```

```python
main_genre = {}
c = 0
for idx,row in df.iterrows():
    # c +=1
    # if c == 20:
    #     break
    genre = row['sp_genres'][1:-1].split(',')[0][1:-1]
    if (genre == ""): continue
    df.at[idx,'sp_genres'] = genre
```

```python
df_eda = df[['popularity', 'sp_genres']]
df_eda = df_eda.groupby(['sp_genres']).mean().sort_values('popularity', ascending=False).head(50)
# df_eda = df_eda['popularity'].sort_values

# print(df_eda['popularity'].sort_values)
df_eda.plot.bar(figsize=(20,7))
plt.xlabel('Genre')
plt.ylabel('Popularity')
plt.title("Spotify genre popularity")
plt.show()
```

```python
df_eda_rym = df[['rym rating', 'sp_genres']]
df_eda_rym = df_eda_rym.groupby(['sp_genres']).mean().sort_values('rym rating', ascending=False).head(50)
# df_eda = df_eda['popularity'].sort_values

# print(df_eda['popularity'].sort_values)
df_eda_rym.plot.bar(figsize=(20,7))
plt.setp(plt.gca(),ylim=(3.5, 4.2))
plt.xlabel('Genre')
plt.ylabel('rym rating')
plt.title("Spotify genre rym rating")
plt.show()
```
