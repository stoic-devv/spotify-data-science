```python
import pandas as pd
import dateutil.parser
from collections import defaultdict
```

```python
df1 = pd.read_csv('../data/temp-data/merge-data-1.csv', sep='~')
df2 = pd.read_csv('../data/temp-data/merge-data-2.csv', sep='~')
df3 = pd.read_csv('../data/temp-data/merge-data-3.csv', sep='~')
df = pd.concat([df1,df2,df3], ignore_index=True)
```

```python
def convert_dt(s):
    return dateutil.parser.parse(s)
    
def rename_to_rym_genres(df):
    df['rym_genres'] = df['genres'][:]
    return df.drop(['genres'], axis=1)

def genre_to_list(s):
    return list(map(lambda x: x.strip(' ').strip('\''), s.strip('[').strip(']').split(',')))
```

```python
df["date"] = df["date"].apply(convert_dt)
df = rename_to_rym_genres(df=df)
```

```python
gdf1 = pd.read_csv('../data/genre-dat-12503.csv')
gdf2 = pd.read_csv('../data/genre-dat-12504-12999.csv')
gdf3 = pd.read_csv('../data/genre-dat-12999-14362.csv')
gdf4 = pd.read_csv('../data/genre-dat-merge-data-2.csv')
gdf5 = pd.read_csv('../data/genre-dat-merge-data-3-434.csv')
```

```python
genre_df = pd.concat([gdf1, gdf2, gdf3, gdf4, gdf5], ignore_index=False)
```

```python
# remove after retrieving remaining data
df = df.loc[:len(genre_df)-1]
```

```python
df['sp_genres'] = pd.Series(genre_df['0'].values)
df['sp_genres'] = df['sp_genres'].apply(genre_to_list)
```

```python
album_count_freq = df.groupby(['album']).count()['id'].to_dict()
new_df = pd.DataFrame()
for k,v in album_count_freq.items():
    temp_df = df[df['album'] == k].sort_values(by='popularity', ascending=False)
    max_len = min(len(temp_df), 6)
    temp_df = temp_df[:max_len]
    new_df = pd.concat([new_df, temp_df])
```

```python
# sort by popularity
# get top 5
genre_freq = defaultdict(int)
for idx, row in new_df.iterrows():
    for g in row['sp_genres']:
        genre_freq[g] += 1
```

```python
df['sp_genres'] = df['sp_genres'].apply(genre_to_list)
genre_freq = defaultdict(int)
for idx, row in df.iterrows():
    for g in row['sp_genres']:
        genre_freq[g] += 1
```

```python
genre_freq
```

```python
# find k popular genres
# map all songs to those genres
# '' -> NA
# if not in popular -> NF
def find_k_popular_genres(d, k=20):
    return set([k for k,v in sorted(d.items(), key= lambda x: x[1],reverse=True)][:k])

pop_genres = find_k_popular_genres(genre_freq, 20)

def apply_genre(x):
    for g in x:
        if g == '':
            return 'NA'
        if g in pop_genres:
            return g
    return 'NF'

genre_series = df['sp_genres'].apply(apply_genre)
```
